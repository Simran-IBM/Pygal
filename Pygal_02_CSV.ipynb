```python
import pygal
import pandas as pd
```

```python
wine_df = pd.read_csv('./datasets/winequality-red.csv',
                                         sep = ',')
```

```python
wine_df.head(2)
```

```python
wine_df.describe()
```

```python
average_df = wine_df.groupby(['quality']).mean()
```

```python
average_df
```

```python
bar_chart = pygal.Bar(height = 100,
                       width = 200,
                    explicity_size = True)
bar_chart.title = 'Average Acidity By Wine Quality'
```

```python
bar_chart.x_title = 'Quality Rating'
bar_chart.x_labels = range(3,9)
```

```python
bar_chart.add('Fixed Acidity', 
               average_df['fixed acidity'])
```
