```python
import pygal
import os
import webbrowser
```

```python
bar_chart = pygal.Bar(height = 300, 
                      width = 500,
                      explicit_size = True)
```

```python
bar_chart.title = "Pet Population in 100 households"
```

```python
bar_chart.add('Dogs', 34)
bar_chart.add('Cats', 24)
bar_chart.add('Hamster', 20)
bar_chart.add('Rabbit', 5)
bar_chart.add('Exotic', 1)
```

```python
bar_chart.render_to_file('pet_bar_chart.svg')
```

```python
def open_in_browser(file_to_open):
    file_path = os.path.realpath(file_to_open)
    file_url = 'file://' + file_path
    chrome_path="C:\\Program Files (x86)\\Google\\Chrome\\Application\\chrome.exe"
    webbrowser.register('chrome', None,webbrowser.BackgroundBrowser(chrome_path))
    webbrowser.get('chrome').open_new_tab(file_url)
```

```python
open_in_browser('pet_bar_chart.svg')
```

```python
horizontal_bar = pygal.HorizontalBar(height = 300, 
                      width = 500,
                      explicit_size = True)
horizontal_bar.title = "Pet Population in 100 households"
```

```python
horizontal_bar.add('Dogs', 34)
horizontal_bar.add('Cats', 24)
horizontal_bar.add('Hamster', 20)
horizontal_bar.add('Rabbit', 5)
```

```python
horizontal_bar.render_to_file('pet_bar_chart_horizontal.svg')
```

```python
open_in_browser('pet_bar_chart_horizontal.svg')
```
